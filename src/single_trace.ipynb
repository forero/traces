```python
r_ini = random.random(3)
r_ini[2] = r_ini[2] - random.random()*10
```

```python
trace_id=0

def likelihood(i_obs, i_model):
    chi_squared = sum((i_obs-i_model)**2)
    return -chi_squared

def my_model(obs_data, r_init):
    trace_prediction = arange(12)
    trace_prediction[0] = obs_data[0]
    trace_prediction[1] = obs_data[1]
    trace_prediction[2] = obs_data[2]

    m_xz = (obs_data[0] - r_init[0])/(obs_data[2] - r_init[2])
    m_yz = (obs_data[1] - r_init[1])/(obs_data[2] - r_init[2])

    trace_prediction[3] = (obs_data[5] - r_init[2])*m_xz + r_init[0]
    trace_prediction[4] = (obs_data[5] - r_init[2])*m_yz + r_init[1]
    trace_prediction[5] = obs_data[5]

    trace_prediction[6] = (obs_data[8] - r_init[2])*m_xz + r_init[0]
    trace_prediction[7] = (obs_data[8] - r_init[2])*m_yz + r_init[1]
    trace_prediction[8] = obs_data[8]

    trace_prediction[9] = (obs_data[11] - r_init[2])*m_xz + r_init[0]
    trace_prediction[10] = (obs_data[11] - r_init[2])*m_yz + r_init[1]
    trace_prediction[11] = obs_data[11]
    return trace_prediction

x_walk = empty((0)) #this is an empty list to keep all the steps
y_walk = empty((0))
z_walk = empty((0))

x_walk = append(x_walk, random.random())
y_walk = append(y_walk, random.random())
z_walk = append(z_walk, random.random())

print x_walk
print y_walk
print z_walk

n_iterations = 100000 #this is the number of iterations I want to make
for i in range(n_iterations):
    x_prime = np.random.normal(x_walk[i], 1.0) 
    y_prime = np.random.normal(y_walk[i], 1.0)
    z_prime = np.random.normal(z_walk[i], 200.0)
    while z_prime > 0.0:
        z_prime = np.random.normal(z_walk[i], 10)

    mod_init = my_model(ob_data[trace_id,:], [x_walk[i],y_walk[i],z_walk[i]])
    mod_prime = my_model(ob_data[trace_id,:], [x_prime,y_prime,z_prime])
    
    alpha = likelihood(ob_data[trace_id,:], mod_prime) - likelihood(ob_data[trace_id,:], mod_init)
    
    if(alpha>0.0):
        x_walk  = append(x_walk,x_prime)
        y_walk  = append(y_walk,y_prime)
        z_walk  = append(z_walk,z_prime)
    else:
        beta = random.random()
        if(beta<exp(alpha)):
            x_walk  = append(x_walk,x_prime)
            y_walk  = append(y_walk,y_prime)
            z_walk  = append(z_walk,z_prime)
        else:
                x_walk  = append(x_walk,x_walk[i])
                y_walk  = append(y_walk,y_walk[i])
                z_walk  = append(z_walk,z_walk[i])
```
